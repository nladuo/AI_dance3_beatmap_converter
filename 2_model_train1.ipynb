### 键型 编码
> 将所有情况弄成一个字典

```python
idx = 0
id2key = {}
key2id = {}
```

```python
key1 = "000000"
if key1 not in key2id:
    key2id[key1] = idx
    id2key[idx] = key1
    idx += 1
```

```python
id2key
```

```python
for y_ in y:
    for y__ in y_:
        key = "".join([str(int(bool(v))) for v in y__.tolist()])  # 这里简化下，就不考虑哪里用长条了。
        if key not in key2id:
            key2id[key] = idx
            id2key[idx] = key
            idx += 1
```

```python
id2key  # 一共22种健型
```

### 构造数据

准备搞一个简单的模型如下，提取一部分时序特征，一部分直接用音频特征，最后再接一个mlp
![](model1.png)

```python
X_data = []
y_data = []

for i in range(len(X)):
    x_tmp = np.zeros
    x_ = np.array(X[i])
    y_ = y[i]
    print(x_.shape, len(y[i]))
    X_data[i, :x_.shape[0], :] = x_
    
    for j, y__ in enumerate(y_):
        idx = key2id["".join([str(v) for v in y__.tolist()])]
        y_data[i, j] = idx
```

### 训练模型

```python
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Embedding

# 编码器模型
encoder_inputs = Input(shape=(None, 100))
encoder_lstm = LSTM(256, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(encoder_inputs)
encoder_states = [state_h, state_c]

# 解码器模型
decoder_inputs = Input(shape=(None,))
decoder_embeddings = Embedding(73, 256)(decoder_inputs)
decoder_lstm = LSTM(256, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_embeddings, initial_state=encoder_states)
decoder_dense = Dense(73, activation='softmax')(decoder_outputs)

# 构建模型
model = Model([encoder_inputs, decoder_inputs], decoder_dense)

model.summary()
```

```python
# 构造解码器输入
decoder_input_data = np.zeros((415, 101))
decoder_input_data[:, 0] = 2  # 使用特殊的开始标记
decoder_input_data[:, 1:] = y_data[:, :-1]  # 目标序列向右移动一位
```

```python
# 编译模型
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 训练模型
model.fit([X_data, decoder_input_data], y_data, epochs=100, batch_size=64)
```

```python
# 初始化解码器输入
decoder_input = np.zeros((1, 1))
decoder_input[0, 0] = 2  # 使用特殊的开始标记

# 存储预测结果
predictions = []

# 逐步预测序列
for _ in range(101):  # 最大序列长度
    output = model.predict([X_test, decoder_input])
    prediction = np.argmax(output[:, -1, :], axis=-1)
    predictions.append(prediction[0])
    
    # 更新解码器输入
    decoder_input = np.zeros((1, 1))
    decoder_input[0, 0] = prediction[0]

y_pred = np.array(predictions)
print(y_pred)
```
